```python
import seaborn as sns
```

```python
sns.get_dataset_names() #all available datasets from seaborn to work with
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

Scatter Plots

```python
sns.scatterplot(x="tip", y="total_bill", data = tips) #we are telling to use tip col as x and tottalbill col as y from tips dataset
```

```python
sns.scatterplot(x="tip", y="total_bill", data = tips, hue = "day", size = "size", palette="YlGnBu") #we cahnged hue as per categorical value from day and we created size of point with tip size and palette is just colors to look good ylGnBu is yellow green blue
```

Histograms and Distributions

```python
sns.histplot(tips["tip"], kde = True, bins = 15) #kde is the line
```

```python
sns.distplot(tips["tip"], kde = True, bins = 15)
```

Bar plot

```python
sns.boxplot(x = "sex", y ="tip", data = tips)
```

```python
sns.boxplot(x = "day", y ="tip", data = tips, hue = "sex")
```

```python
sns.stripplot(x = "day", y ="tip", data = tips, hue = "sex")
```

```python
sns.stripplot(x = "day", y ="tip", data = tips, hue = "sex", dodge = True) #dodge is speartion between datapoints
```

Joint Plot -> just a combination of scatter and dist plot

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips)
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "reg") #add reg to get a regression line
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "kde") #to get a more of a 3d plot
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "kde", shade = True, cmap = "YlGnBu")
```

```python
#we don't use shade in hex
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "hex")
```
